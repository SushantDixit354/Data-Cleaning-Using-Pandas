## Change Data type 

### quantity column Object -> float16

```python
index = cafe[cafe['quantity'].isna()].index
cafe.loc[index,'quantity'] = 'NaN'
```

```python
index = cafe[cafe['quantity'].str.isalpha()].index
cafe.loc[index,'quantity'] = np.nan
```

```python
cafe['quantity'].isna().sum()
```

```python
cafe['quantity'] = cafe['quantity'].astype('float16')
```

## Filling Missing Values on Numerical Columns

### total_spent  = quantity*price_per_unit

```python
tempdf = cafe[cafe['total_spent'].isna()]
```

```python
total_spent = tempdf['quantity']*tempdf['price_per_unit']
```

```python
index = tempdf.index
```

```python
cafe.loc[index,'total_spent'] = total_spent
```

#### NULL value is reduced form 503 to 40 (reduced by 92%)

```python
cafe['total_spent'].isna().sum() 
```

### quantity = total_spent/price_per_unit

```python
cafe['quantity'].isna().sum() # total null values 479
```

```python
tempdf = cafe[cafe['quantity'].isna()]
```

```python
quantity = tempdf['total_spent']/tempdf['price_per_unit']
```

```python
index = tempdf.index
```

```python
cafe.loc[index,'quantity'] = quantity
```

```python
cafe['quantity'].isna().sum() # total null values left 38
```

### price_per_unit = total_spent/quantity

```python
index = tempdf.index
```

```python
cafe.loc[index,'price_per_unit'] = price_per_unit 
```
